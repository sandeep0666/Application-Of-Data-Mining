```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = "/content/airplane_price_dataset.csv"
df = pd.read_csv(file_path)
```

```python
df_encoded = pd.get_dummies(df, columns=['Motor Türü', 'Satış Bölgesi'], drop_first=True)
```

```python
label_encoder = LabelEncoder()
df_encoded['Model'] = label_encoder.fit_transform(df_encoded['Model'])
```

```python
current_year = 2025
df_encoded['Computed Age'] = current_year - df_encoded['Üretim Yılı']
```

```python
avg_price_by_motor = df.groupby('Motor Türü')['Fiyat ($)'].mean()
print("Average Price by Motor Type:\n", avg_price_by_motor)
```

```python
pivot_table = df.pivot_table(values='Fiyat ($)', index='Satış Bölgesi', columns='Motor Sayısı', aggfunc='mean')
print("\nPivot Table (Avg. Price by Satış Bölgesi & Motor Sayısı):\n", pivot_table)
```
