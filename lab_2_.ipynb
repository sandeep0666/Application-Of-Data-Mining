```python
df=pd.DataFrame({
    'math':[85,79,89,46,65],
    'science':[48,62,76,48,45],
    'english':[75,59,45,23,74]
  })
```

```python
correlation_matrix=df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('correlation heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['math'], label='math', color='blue')
plt.title('Line Plot: Age Across Index')
plt.xlabel('Index')
plt.ylabel('Age')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
math_counts = df['math'].value_counts()
labels = [str(score) for score in math_counts.index]
math_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=labels,
    colors=['red', 'green', 'blue', 'yellow', 'orange'],
    title='Pie Chart: Math Score Distribution'
)
plt.ylabel('')
plt.show()
```
